```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sea
```

# 1. Reading the dataset

```python
df = pd.read_csv('Churn_Modelling.csv')
print(df)
```

# 3.1. Univariate Ananlysis

```python
count = 0
for col in df.columns[3:]:
    plt.figure(count)
    count+=1
    if col.title() in ['Creditscore', 'Balance', 'Tenure', 'Estimatedsalary']:
        df[col].plot(kind = 'density')
    else:
        sea.histplot(df[col], ax = plt.gca())
        plt.ylabel('Count')
    plt.xlabel(col.title())
```

# 3.2 Bivariate Analysis

```python
sea.scatterplot(x = df['CreditScore'], y = df['Balance'])
plt.figure(count+1)
sea.scatterplot(x = df['Balance'], y=df['EstimatedSalary'])
plt.figure(count+2)
sea.scatterplot(x = df['CreditScore'], y = df['EstimatedSalary'])
plt.figure(count+3)
```

```python
print('Correlation between Credit score and balance '+ str(df['CreditScore'].corr(df['Balance'])))
print('Correlation between balance and Estimated salary '+ str(df['Balance'].corr(df['EstimatedSalary'])))
print('Correlation between Credit score and Estimated salary '+ str(df['CreditScore'].corr(df['EstimatedSalary'])))
print('Covariance between Credit score and balance '+ str(df['CreditScore'].cov(df['Balance'])))
print('Covariance between balance and Estimated salary '+ str(df['Balance'].cov(df['EstimatedSalary'])))
print('Covariance between Credit score and Estimated salary '+ str(df['CreditScore'].cov(df['EstimatedSalary'])))
```
